### Cluster

```python
numeric_feats=['dias_primera_compra', 'dias_ultima_compra', 'ventas',
       'frecuencia', 'monto']
```

```python
X=df010[numeric_feats]
```

```python
########################
#   Estandarizacion
########################


X_std = DataFrame(StandardScaler().fit_transform(X), columns=numeric_feats)
```

```python
########################
#   Hierarchy
########################

X_train, X_test = train_test_split(X_std, test_size=0.9995, random_state=42)

len(X_train)

linked = linkage(X_train, 'single')

labelList = range(1, 11)

plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

```python
########################
#        Elbow
########################


X=X_train

# k means determine k
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    kmeanModel.fit(X)
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])

# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
for n_clusters in range(3,10):
    kmeans = KMeans(init='k-means++', n_clusters = n_clusters, n_init=30)
    kmeans.fit(X)
    clusters = kmeans.predict(X)
    silhouette_avg = silhouette_score(X, clusters)
    print("For n_clusters =", n_clusters, "The average silhouette_score is :", silhouette_avg)
```

```python
plt.scatter(X['ventas'],
            X['frecuencia']);
```

```python
kmeans = KMeans(init='k-means++', n_clusters = 4, n_init=30)
kmeans.fit(X)
clusters = kmeans.predict(X)
    
plt.scatter(X['ventas'],
            X['frecuencia'], c=clusters, s=50, cmap='viridis')
```

```python
plt.scatter(X['dias_primera_compra'],
            X['frecuencia'], c=clusters, s=50, cmap='viridis')
```

```python
X['cluster']=pd.Series(clusters, index=X.index)
```
